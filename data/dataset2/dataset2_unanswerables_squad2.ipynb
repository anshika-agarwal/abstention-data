```python
# DATASET 2: UNANSWERABLES FROM SQUAD 2.0 (TRAIN)

import json, random

NO_ANS = "<NO-ANSWER>"
TARGET_SIZE = 3000
SEED = 42

def load_squad_unanswerable(path):
    obj = json.load(open(path))
    rows = []
    for article in obj["data"]:
        for para in article["paragraphs"]:
            for qa in para["qas"]:
                if qa.get("is_impossible", False):
                    rows.append({
                        "id": qa["id"],
                        "question": qa["question"].strip(),
                        "label": NO_ANS,
                        # optional metadata
                        "title": article.get("title", ""),
                        "context": para.get("context", ""),
                    })
    return rows

rows = load_squad_unanswerable("train-v2.0.json")
print("Total unanswerables available:", len(rows))

random.seed(SEED)
sampled = random.sample(rows, TARGET_SIZE)

def write_jsonl(path, rows):
    with open(path, "w") as f:
        for r in rows:
            f.write(json.dumps({
                "id": r["id"],
                "input": r["question"],
                "output": r["label"]
            }) + "\n")

write_jsonl("dataset2_unanswerable_train_3000.jsonl", sampled)

print("Wrote:", len(sampled))
```
