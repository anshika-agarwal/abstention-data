```python
import json

# Convert metadata_json strings to dictionaries
def parse_metadata(metadata_str):
    try:
        if isinstance(metadata_str, str):
            return json.loads(metadata_str)
        elif isinstance(metadata_str, dict):
            return metadata_str
        else:
            return {}
    except:
        return {}

# Parse all metadata_json values
df_unanswerable['metadata_parsed'] = df_unanswerable['metadata_json'].apply(parse_metadata)

# Get all unique keys across all dictionaries
all_keys = set()
for metadata_dict in df_unanswerable['metadata_parsed']:
    if isinstance(metadata_dict, dict):
        all_keys.update(metadata_dict.keys())

print(f"Unique keys found: {sorted(all_keys)}")
print(f"Total unique keys: {len(all_keys)}")
```

```python
import json
import re

def extract_dataset_from_metadata(metadata_json):
    """Extract dataset name from metadata_json keys"""
    try:
        # Parse metadata
        if isinstance(metadata_json, str):
            metadata = json.loads(metadata_json)
        elif isinstance(metadata_json, dict):
            metadata = metadata_json
        else:
            return "unknown"
        
        if not isinstance(metadata, dict):
            return "unknown"
        
        # Get all keys
        keys = list(metadata.keys())
        
        # Dataset mapping - prefix to dataset name
        dataset_map = {
            'ALCUNA': 'ALCUNA',
            'BBQ': 'BBQ',
            'BigBenchKnownUnknowns': 'BigBenchKnownUnknowns',
            'CoCoNot': 'CoCoNot',
            'KUQ': 'KUQ',
            'MediQ': 'MediQ',
            'MoralChoice': 'MoralChoice',
            'QAQA': 'QAQA',
            'QASPER': 'QASPER',
            'SQuAD2.0': 'SQuAD2.0',
            'SituatedQA': 'SituatedQA',
            'UMWP': 'UMWP'
        }
        
        # Find dataset by checking key prefixes
        found_datasets = []
        for key in keys:
            for prefix, dataset_name in dataset_map.items():
                if key.startswith(prefix + '_'):
                    found_datasets.append(dataset_name)
                    break
        
        # Return first found dataset, or unknown if none
        if found_datasets:
            return found_datasets[0]  # Take first match
        else:
            return "unknown"
    
    except Exception as e:
        return "unknown"

# Add dataset column
df_unanswerable['dataset_from_metadata'] = df_unanswerable['metadata_json'].apply(extract_dataset_from_metadata)

# Show results
print("Dataset distribution:")
print(df_unanswerable['dataset_from_metadata'].value_counts())
print(f"\nTotal: {len(df_unanswerable)}")
print(f"Unknown: {(df_unanswerable['dataset_from_metadata'] == 'unknown').sum()}")
```

```python
import json
from pathlib import Path
import random
import shutil
from datetime import datetime
from collections import Counter, defaultdict

# File paths
abstention_bench_path = Path(r"C:\Users\arjun\OneDrive\Documents\Stanford\Coterm\Winter\CS525\abstention-data\data\abstentionbench\abstention_bench_dataset.jsonl")
dataset4_path = Path(r"C:\Users\arjun\OneDrive\Documents\Stanford\Coterm\Winter\CS525\abstention-data\data\dataset4.jsonl")

# Create backups
print("Creating backups...")
shutil.copy2(abstention_bench_path, abstention_bench_path.with_suffix('.jsonl.backup'))
shutil.copy2(dataset4_path, dataset4_path.with_suffix('.jsonl.backup'))
print("  ✓ Backups created")

# Step 1: Load dataset4.jsonl
print("\nLoading dataset4.jsonl...")
dataset4_examples = []
with open(dataset4_path, 'r', encoding='utf-8') as f:
    dataset4_examples = [json.loads(line) for line in f]
print(f"  Total examples: {len(dataset4_examples):,}")

# Separate answerable and unanswerable
answerable_examples = [ex for ex in dataset4_examples if ex.get('output') != '<NO-ANSWER>']
unanswerable_examples = [ex for ex in dataset4_examples if ex.get('output') == '<NO-ANSWER>']

print(f"  Answerable: {len(answerable_examples):,}")
print(f"  Unanswerable: {len(unanswerable_examples):,}")

# Step 2: Reduce unanswerable to 250
random.seed(42)
num_to_keep = min(250, len(unanswerable_examples))
kept_unanswerable = random.sample(unanswerable_examples, num_to_keep)
print(f"\nReduced unanswerable to: {len(kept_unanswerable)} examples")

# Step 3: Load abstention_bench_dataset.jsonl
print("\nLoading abstention_bench_dataset.jsonl...")
abstention_bench_examples = []
with open(abstention_bench_path, 'r', encoding='utf-8') as f:
    abstention_bench_examples = [json.loads(line) for line in f]
print(f"  Total examples: {len(abstention_bench_examples):,}")

# Filter to available examples (not already used)
available_examples = [ex for ex in abstention_bench_examples 
                     if not ex.get('used_in_dataset4', False)]
print(f"  Available examples: {len(available_examples):,}")

# Filter to target scenarios
target_scenarios = ['Answer Unknown', 'False Premise', 'Subjective']
filtered_examples = [ex for ex in available_examples 
                    if ex.get('scenario') in target_scenarios]

print(f"\nFiltered to target scenarios: {len(filtered_examples):,} examples")
scenario_counts = Counter([ex.get('scenario') for ex in filtered_examples])
for scenario, count in scenario_counts.items():
    print(f"  {scenario}: {count:,} examples")

# Step 4: Group by scenario and sample 250 from each
scenario_groups = defaultdict(list)
for ex in filtered_examples:
    scenario = ex.get('scenario')
    scenario_groups[scenario].append(ex)

samples_to_add = []
sample_ids = set()
random.seed(42)

for scenario in target_scenarios:
    if scenario not in scenario_groups:
        print(f"  Warning: {scenario} not found in available examples")
        continue
        
    examples = scenario_groups[scenario]
    num_to_sample = min(250, len(examples))
    
    sampled = random.sample(examples, num_to_sample)
    
    # Convert to dataset4 format
    for ex in sampled:
        # Create new example in dataset4 format
        new_ex = {
            'id': ex.get('id', f"abstention_bench_{ex.get('id', 'unknown')}"),
            'input': ex.get('question', ex.get('input', '')),
            'output': ex.get('output', '<NO-ANSWER>'),
            'split': 'dev'
        }
        samples_to_add.append(new_ex)
        sample_ids.add(ex.get('id', ''))
    
    print(f"  {scenario}: sampled {num_to_sample} examples")

print(f"\nTotal sampled from AbstentionBench: {len(samples_to_add)} examples")

# Step 5: Mark sampled examples in abstention_bench_dataset.jsonl
timestamp = datetime.now().isoformat()
for ex in abstention_bench_examples:
    if ex.get('id', '') in sample_ids:
        ex['used_in_dataset4'] = True
        ex['added_to_dataset4_date'] = timestamp

# Save updated abstention_bench
with open(abstention_bench_path, 'w', encoding='utf-8') as f:
    for ex in abstention_bench_examples:
        f.write(json.dumps(ex, ensure_ascii=False) + '\n')
print(f"\n✓ Marked examples in abstention_bench_dataset.jsonl")

# Step 6: Combine all examples for new dataset4
# Keep all answerable + 250 unanswerable from original + 750 from AbstentionBench
new_dataset4 = answerable_examples + kept_unanswerable + samples_to_add
random.shuffle(new_dataset4)

# Save new dataset4.jsonl
with open(dataset4_path, 'w', encoding='utf-8') as f:
    for example in new_dataset4:
        f.write(json.dumps(example, ensure_ascii=False) + '\n')

print(f"\n✓ Updated dataset4.jsonl")
print(f"\n{'='*60}")
```

```python
import json
from pathlib import Path
from collections import defaultdict
import re

# File paths
abstention_bench_path = Path(r"C:\Users\arjun\OneDrive\Documents\Stanford\Coterm\Winter\CS525\abstention-data\data\abstentionbench\abstention_bench_dataset.jsonl")
dataset4_path = Path(r"C:\Users\arjun\OneDrive\Documents\Stanford\Coterm\Winter\CS525\abstention-data\data\dataset4.jsonl")

# Step 1: Load AbstentionBench and create normalized question -> scenario mapping
print("Loading abstention_bench_dataset.jsonl...")
question_to_scenario = {}

def normalize_text(text):
    """Normalize text for comparison (remove extra whitespace)"""
    return ' '.join(text.strip().split())

with open(abstention_bench_path, 'r', encoding='utf-8') as f:
    for line in f:
        ex = json.loads(line)
        question = ex.get('question', ex.get('input', '')).strip()
        scenario = ex.get('scenario', '')
        if question and scenario:
            normalized_question = normalize_text(question)
            question_to_scenario[normalized_question] = scenario

print(f"  Mapped {len(question_to_scenario)} questions to scenarios")

# Step 2: Load dataset4
print("\nLoading dataset4.jsonl...")
dataset4_examples = []
with open(dataset4_path, 'r', encoding='utf-8') as f:
    for line in f:
        dataset4_examples.append(json.loads(line))

print(f"  Loaded {len(dataset4_examples)} examples")

# Step 3: Add unanswerable_type by matching questions
print("\nMatching questions and adding unanswerable_type...")
matched_count = 0
unmatched_count = 0

for ex in dataset4_examples:
    output = ex.get('output', '')
    
    if output == '<NO-ANSWER>':
        input_text = ex.get('input', '').strip()
        normalized_input = normalize_text(input_text)
        
        if normalized_input in question_to_scenario:
            scenario = question_to_scenario[normalized_input]
            if scenario in ['Answer Unknown', 'False Premise', 'Subjective']:
                ex['unanswerable_type'] = scenario
            else:
                ex['unanswerable_type'] = scenario
            matched_count += 1
        else:
            ex['unanswerable_type'] = 'Underspecified Context'
            unmatched_count += 1
    else:
        ex['unanswerable_type'] = None

print(f"  Matched: {matched_count} (from AbstentionBench)")
print(f"  Unmatched: {unmatched_count} (Underspecified Context)")

# Step 4: Verify distribution
print("\nUnanswerable type distribution:")
unanswerable_types = defaultdict(int)
for ex in dataset4_examples:
    if ex.get('output') == '<NO-ANSWER>':
        unanswerable_type = ex.get('unanswerable_type', 'Unknown')
        unanswerable_types[unanswerable_type] += 1

for unanswerable_type, count in sorted(unanswerable_types.items()):
    print(f"  {unanswerable_type}: {count}")

# Step 5: Save updated dataset4
print(f"\nSaving updated dataset4.jsonl...")
with open(dataset4_path, 'w', encoding='utf-8') as f:
    for example in dataset4_examples:
        f.write(json.dumps(example, ensure_ascii=False) + '\n')

print(f"  ✓ Saved {len(dataset4_examples)} examples with unanswerable_type column")
```
